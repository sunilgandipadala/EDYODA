#You have to design a Food Ordering app for a restaurant

###The application will have a log-in for admin and users to log-in

-------------------------------- Admin ----------------------------------

➡️ Admin will have the following functionalities: ⬅️


👉 1. Add new food items. Food Item will have the following details:

        🔴 FoodID //It should be generated automatically by the application.
        🔴 Name
        🔴 Quantity. For eg, 100ml, 250gm, 4pieces etc
        🔴 Price
        🔴 Discount
        🔴 Stock. Amount left in stock in the restaurant.

👉 2. Edit food items using FoodID.

👉 3. View the list of all food items.

👉 4. Remove a food item from the menu using FoodID.

--------------------------------- User ----------------------------------

➡️ The user will have the following functionalities: ⬅️

👉 1. Register on the application. Following to be entered for registration:

        🔴 Full Name
        🔴 Phone Number
        🔴 Email
        🔴 Address
        🔴 Password

👉 2. Log in to the application

👉 3. The user will see 3 options:

        🔴 Place New Order
        🔴 Order History
        🔴 Update Profile

👉 4. Place New Order: The user can place a new order at the restaurant.

        🔵 Show list of food. The list item should as follows:
            🔴 Tandoori Chicken (4 pieces) [INR 240]
            🔴 Vegan Burger (1 Piece) [INR 320]
            🔴 Truffle Cake (500gm) [INR 900]

👉 5. Users should be able to select food by entering an array of numbers. For example, if the user wants to order Vegan Burger and Truffle Cake they should enter [2, 3]

👉 6. Once the items are selected user should see the list of all the items selected. The user will also get an option to place an order.

👉 7. Order History should show a list of all the previous orders

👉 8. Update Profile: the user should be able to update their profile.

```python
class Restaurant:
  counter = 1
  menu_list = {
      1:{"Name":"Idly","quantity":"100","price":10,"discount":2,"stock":7}
  }
  user_list = {}

class Admin(Restaurant):

  def addItem(self,FoodID,Name,Quantity,Price,Discount,Stock):
    if FoodID not in self.menu_list.keys():
      self.menu_list[FoodID] = {
          "Name": Name,
          "quantity":Quantity,
          "price":Price,
          "discount":Discount,
          "stock" : Stock
      }
      print("Item added.. successfuly..")
    else:
      print("Item already there, will update the quantity,discount and stock")
      self.menu_list[FoodID]["price"] += Price
      self.menu_list[FoodID]["stock"] += Stock
      self.menu_list[FoodID]["discount"] = Discount
      print("Item details were modified...")
      

  def editFoodItem(self,FoodID,Name,Quantity,Price,Discount,Stock):
    if FoodID not in self.menu_list.keys():
      print("There is no item in the menu, first add the item to the menu...")
      return 
    else:
      self.menu_list[FoodID]["Name"] = Name
      self.menu_list[FoodID]["quantity"] = Quantity
      self.menu_list[FoodID]["price"] = Price
      self.menu_list[FoodID]["stock"] = Stock
      self.menu_list[FoodID]["discount"] = Discount
      print("Food Item edited succesfully..")
  def removeFoodItem(self,FoodID):
    if FoodID not in self.menu_list.keys():
      print("There is no food item in the menu to remove..")
    else:
      del self.menu_list[FoodID]
      print("Item Removed...")
  def viewMenu(self):
    all_items = []
    for i in self.menu_list.keys():
      item = self.menu_list[i]["Name"]
      all_items.append(item)
    print("All the items in the menu are \n",all_items)
  
## User Class
class User(Restaurant):
  order_list={}
  logged_in = 0
    
  def register(self,fullName,phNo,email,address,password):
    self.current_user = 0
    if email not in self.user_list.keys():
      self.user_list[email]={
          "fullName":fullName,
          "phNo":phNo,
          "address":address,
          "password":password
      }
      print("Registration Successful..Login in to proceed")
    else:
      print("You are already registered, Login to continue..")
  def login(self, email,password):
    if email in self.user_list.keys():
      if self.user_list[email]["password"] == password:
        self.current_user = email
        self.logged_in = 1
        print("Login Succesfful")
      else:
        print("Enter valid details..")
    else:
      print("You need to register .. to login")
  def placeOrder(self,FoodIDs):
    self.order_list[self.current_user]={}
    print("These are the items, you have ordered..,please confirm your order")
    for id in FoodIDs:
      if id in self.menu_list.keys():
        order = self.menu_list[id]["Name"],self.menu_list[id]["quantity"],self.menu_list[id]["price"]
        self.order_list[self.current_user][id]=order
        print(order)
      else:
        print("item is not available..")
    
    print("Do you want to confirm your order.. ?\npress 1 or 2\n1- Yes \t 2- No")
    opt = int(input())
    if opt ==1:
      print("Your order confirmed..")
    else:
      print("Your order cancelled..re enter your order..")
      foodids= [int(i) for i in input().split()]
      self.placeOrder(foodids)

  def orderHistory(self):
    print(self.current_user+" You orders")
    print(self.order_list[self.current_user])
  
  def updateProfile(self,fullName,phNo,email,address,password):
    if email in self.user_list.keys():
      self.user_list[email]={
          "fullName":fullName,
          "phNo":phNo,
          "address":address,
          "password":password
      }
    else:
      print("You need to register...")
  
```

```python
print("\t\t----------------Welcome to Edyoda Food Delivery App-----------------------")
while True:
  print("\nWhat is your role..?")
  print("\t 1 - Admin")
  print("\t 2 - User")
  print("Choose your role...")
  role = int(input())
  if role == 1:
    print("\nHellooo.. Admin.. welcome to your page")
    admin= Admin()
    while True:
      print("\nThe following are the functionalities you can do... ")
      print("1- Add New Food Item\n2- Edit Food Item\n3- View List of Food Items\n4- Remove Food Item \n5- Exit()\n")
      choice = int(input("Enter your choice"))
      if choice == 1:
        print("\t\t\t Enter the new food item details..")

        id = len(admin.menu_list)+1
        name = input("name: ")
        quant = int(input("quantity: "))
        price = int(input("price: "))
        disc = int(input("discount: "))
        stock = int(input("stock: "))
        admin.addItem(id,name,quant,price,disc,stock)
      elif choice == 2:
        print("\t\t\t Enter the food item details to edit")
        id = int(input("Food ID: "))
        name = input("name: ")
        quant = int(input("quantity: "))
        price = int(input("price: "))
        disc = int(input("discount: "))
        stock = int(input("stock: "))
        admin.editFoodItem(id,name,quant,price,disc,stock)
      elif choice == 3:
        admin.viewMenu()
      elif choice == 4:
        print("\t\t\t Enter the food item want to remove..")
        id = int(input("Food ID:"))
        admin.removeFoodItem(id)
      else:
        print("You are logged out..\n")
        break

  elif role == 2:
    print("Hellooo.. Welcome..\n")
    print("Register / Login to continue...")
    user = User()
    while user.logged_in !=1 :
      print("\n1 -Register\n2 -Login\n3 -Close Application")
      choice = int(input("Register|Login.."))
      if choice == 1:
        fullName = input("Full Name:")
        phNo = input("Phone NO:")
        email = input("Email:")
        address = input("Address")
        password = input("Password")
        user.register(fullName,phNo,email,address,password)
      elif choice == 2:
        email = input("Email:")
        password = input("Password:")
        user.login(email,password)
      else:
        exit("Going back to home page...\n")
    while True:
      print("\nThe following are the functionalities you can do... ")
      print("\n1- Place New Order\n2- Order History\n3- Update profile\n4- Exit()\n")
      choice = int(input("Enter your choice"))
      if choice == 1:
        print("The list items are:")
        for i in user.menu_list.keys():
          print("\t\t\t",user.menu_list[i]["Name"],"("+str(user.menu_list[i]["quantity"])+"pieces)","[ INR "+str(user.menu_list[i]["price"])+"]")
        foodids= [int(i) for i in input("Enter the list of foodids: ").split()]
        user.placeOrder(foodids)
      elif choice == 2:
        user.orderHistory()
      elif choice == 3:
        fullName = input("Full Name:")
        phNo = input("Phone NO:")
        email = input("Email:")
        address = input("Address")
        password = input("Password")
        user.updateProfile(fullName,phNo,email,address,password)
      else:
        print("You are logged out..\n")
        user.logged_in =0
        break
  else:
    print("Application closed...\n")







  
```
